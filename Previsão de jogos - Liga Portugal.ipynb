```python
dados = pd.read_csv('jogos.csv')
```

```python
from pycaret.classification import *
```

```python
s = setup(dados, target = 'Golos Fora', fold = 8)
```

```python
best_model = compare_models()
```

```python
tuned_data = tune_model(best_model, choose_better = True)
```

```python
evaluate_model(tuned_data)
```

```python
predict_model(tuned_data);
```

```python
final_model = finalize_model(tuned_data)
```

```python
unseen_predictions = predict_model(final_model, data=dados)
unseen_predictions
```
